```python
def getvar(L, a1, c0=0, c1=1, dx=30):
    dh = dx/np.sqrt(np.pi)
    c = c0 + c1
    out = np.zeros_like(L)
    idx = L <= dh
    out[idx] = c
    idx = (L > dh) & (L < a1)
    out[idx] = c0*(dh**2/L[idx]**2) + c1*(1-(L[idx]/a1)+(1/5)*(L[idx]/a1)**3)
    idx = L > a1
    out[idx] = c0*(dh**2/L[idx]**2) + (1/5)*c1*(a1**2/L[idx]**2)
    return out
    
```

```python
L = np.linspace(0,1E4,1000)
```

```python
v = getvar(L, 500)
sigma = np.sqrt(v)
```

```python
f, ax = plt.subplots()
ax.semilogx(L, sigma, label='$a_1=%i\ m$' % a1)
#ax.semilogx(L, v, label='$a_1=%i\ m$' % a1)
ax.axvline(a1, lw=0.5, ls='--')
ax.set_ylim(0,1.0)
ax.legend()
```

```python
dM = dh*A*rho
dM
```

```python
np.sqrt((dh_sigma/dh)**2+(A_sigma/A)**2+(rho_sigma/rho)**2)
```

```python
np.sqrt((dh_sigma/dh)**2+(A_sigma/A)**2+(rho_sigma/rho)**2)*np.abs(dM)
```

```python
dV_sigma = np.sqrt((dh_sigma*A)**2 + (A_sigma*dh)**2)
dV_sigma
```

```python
dM_sigma = np.sqrt((dV_sigma*rho)**2 + (rho_sigma*(dh*A))**2)
dM_sigma
```
